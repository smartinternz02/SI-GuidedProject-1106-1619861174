```python
import seaborn as sns
```

```python
iris=sns.load_dataset('iris')
```

```python
iris
```

```python
sns.scatterplot(x="sepal_width",y="sepal_length",data=iris)
```

```python
sns.lineplot(x="petal_width",y="petal_length",data=iris)
```

```python
sns.distplot(iris["sepal_length"])
```

```python
sns.barplot(x="petal_length",y="species",data=iris)
```

```python
sns.boxplot(x="species",y="petal_width",data=iris)
```

```python
sns.catplot(x="species",y="sepal_width",data=iris)
```

```python
sns.pairplot(iris)
```

```python
sns.heatmap(iris.corr(),annot=True)
```
